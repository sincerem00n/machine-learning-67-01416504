## Generate Linear Data with noise and use Linear Regression model to train the model. Find MAE, MSE and RMSE

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(21)
x = np.random.rand(100, 1)
y = 2 * x + 1 + np.random.normal(0, 1, x.shape)

plt.scatter(x, y)
plt.show()
```

```python
# Linear regression
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x, y)
print(f'Intercept: {model.intercept_}')
```

```python
# Find MAE, MSE and RMSE
from sklearn.metrics import mean_absolute_error, mean_squared_error
y_pred = model.predict(x)
mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)
print(f'MSE: {mse}, MAE: {mae}, RMSE: {rmse}')
```

```python
# plot the regression line
plt.scatter(x, y)
plt.plot(x, y_pred, color='red')
plt.legend(['Data', 'Regression line'])
# add the mse, mae and rmse to the plot
plt.text(0.4, -0.5, f'MSE: {mse:.2f}, MAE: {mae:.2f}, RMSE: {rmse:.2f}', fontsize=10)
plt.show()
```
